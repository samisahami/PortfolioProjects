```python
movies_counted = movies_genre[movies_genre['vote_count'] >= 50]

movies_counted.corr(method = 'spearman')
```

```python
import seaborn as sns

sns.regplot(x = 'vote_average', y= 'revenue', data = movies_counted, line_kws = {"color": 'red'})
```
